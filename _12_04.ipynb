```python
# Yapay Zekanın Geleceği 
# url : yapay-zekanın-geleceği

title = "Yapay Zekanın Geleceği".lower()
words = title.split()

#key - value 
turkish_map = {
    'Ç':'c',
    'ğ':'g',
    'ı':'i'
}

clean_list = []

for word in words:
    new_word = ""
   
    for char in word :
        if char in turkish_map:
            new_word += turkish_map[char]  
        else: 
            new_word += char  
    
    clean_list.append(new_word)
    
url = "-".join(clean_list)
 
print(title)
print (url)

```
