```python
# String Metotları 

names = ["  Hava  ","  Adem  ","  Şirin  ","  Yavuz  "]

for name in names: 
    print(len (name))
```

```python
for name in names:
    print(len (name.strip()))
```

```python
for name in names:
    print(name, len (name.strip()))
```

```python
for name in names:
    print (f"{name.strip():<10}{len (name.strip())}")
```

```python
for name in names:
    print (f"{name.strip():<10}{len (name.strip()):.1f}")   #.1f araya nokta koydu ondalıklı yaptı 
```

```python
for name in names:
    print (f"{name.strip():<10}{len (name.strip())}\t {len(name)}")
```

```python
pdfs = []

files = ["rapor.pdf","resim.jpg","manzara.png","dokuman.pdf"]

for file in files:
    new_file = file.split(sep='.')
    print(new_file)
    if (new_file[1]=="pdf"):
        pdfs.append(file)
print(pdfs)
```

```python
pdfs = []
for file in files:
    if file.endswith("pdf"):
        pdfs.append(file)
print(pdfs)
```
