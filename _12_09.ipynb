```python
# Sezar Şifreleme

text = input (" Lütfen Şifrelenicek Metni Giriniz: ")
shift = int (input ("Kaç Harf Kaydırılsın (örn:3)"))

text = text.upper()
encryted_text = ""

for char in text:
    if char.isalpha():
        new_char_code = ord(char) + shift
        if new_char_code>ord ('Z'):
            new_char_code= ord('A') + (new_char_code-ord('Z')-1)
        encryted_text += chr(new_char_code)
    else:
        encryted_text += char

print (f"Girilen İfade: {text}")
print (f"Şifrelenmiş İfade: {encryted_text}")
```
