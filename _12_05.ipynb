```python
exp = "Bilimsel Araştırmalar"
harf = 'a'
sayac = 0

for char in exp:
    if char.lower()==harf:
        sayac+=1
        
print (f"{harf } harfi {sayac } kez ilgili ifade de geçiyor.")
```
