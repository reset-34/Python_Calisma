```python
while True :
    try:
        n = int (input("Lütfen Geçerli Bir Tam Sayı giriniz : "))
        if n>0:
            break
        else:
            print ("Lütfen Geçerli Bir Tam Sayı giriniz !")
    except:
        print("Hata !")
toplam = 0
for i in range (1, n+1):
    toplam += i
print (f"1'den {n} 'e kadar olan sayıların toplamı: {toplam}")
        
```
