```python
word1 = "Elma"
word2 = "amel"

if sorted(word1.lower()) == sorted(word2.lower()):     #küçük harfe çevirdik 
    print ("Bu iki kelime anagramdır")
else:
    print ("Bu iki kelime anagram değildir.")
```
