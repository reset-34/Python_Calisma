```python
n = int (input("n sayısını giriniz."))
if n<=1:
    print (f"{n} asal sayı değildir ! ")
else:
    asal = True
    for i in range (2, n ):
        if n % i == 0:
            asal = False
            break
    if asal:
        print(f"{n} bir asal sayıdır.")
    else:
        print(f"{n} bir asal sayı değildir.")   
        
    
```
