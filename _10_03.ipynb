```python
import random
n = int (input("Sayır Sayısını Giriniz: "))
m = int (input("Sütun Sayısını giriniz: "))

matris = []
for i in range (n):
    satir_listesi =[]
    for j in range (m):
        satir_listesi.append(random.randint(0,2))
        matris.append(satir_listesi)

print (matris)
```

```python
for i in range (len (matris)):
    for j in range (len (matris [0])):
        print (f"{matris[i][j]:4d}",end=' ')
    print (" ")
```

```python
for i in range (len (matris)):
    for j in range (len (matris [0])):
        print (f"{matris [i] [j]}")
```

```python
# transpoz 

transpoze = []
for j in range (len(matris[0])):
    yeni_satır = []
    for i in range (len (matris)):
        yeni_satır.append(matris[i][j])
    transpoze.append(yeni_satır)

```

```python
for i in range (len (transpoze)):
    for j in range (len (transpoze [0])):
        print (f"{transpoze[i][j]:4d}",end=' ')
    print (" ")
```
