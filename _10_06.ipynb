```python
n = int (input ("Satır Sayısını Giriniz: "))
m = int (input ("Sütun Sayısını Giriniz: "))

m1 = [[0]*m for _ in range (n)]   # i kullanmadığımız da _ ı i yerine kullanıyoruz
m2 = [[0]*m for _ in range (n)]
m3 = [[0]*m for _ in range (n)]

#değer atama / gezinme / yeni atama 

for i in range (n):
    for j in range (m):
        m1 [i][j] = (i+1) + (j+1)
        m2 [i][j] = (i+1) * (j+1)
        m3 [i][j] = m1[i][j] + m2[i][j]      # matrislerin toplama işlemini yaptırmış olduk 
        
# print (m1) i formatlı çıkarmak için aşağıdaki kod u uyguladık 
for i in range (n):
    for j in range (m):
        print (f"{m1[i][j]:4d}",end= ' ')
    print(" ")

print (m2)
print (m3)

```
