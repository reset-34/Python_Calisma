```python
def square (number):
    return number**2

print(square(4))

numbers = [1,3,5,7]
for number in numbers:
    print (square(number))
```

```python
def is_even(n):
    return n%2==0

def check_number(number):
    if is_even(number):
        print(f"{number} çifttir. ")
    else:
        print(f"{number} tektir. ")
for number in numbers:
    check_number(number)
```

```python
def add (*numbers):
    total = 0
    for n in numbers:
        total += n
    return total


print (add(5))
print (add(5,3))
print (add(5,3,9))
print (add(5,3,9,10))
```

```python
def print_details(**info):
    for key, value in info.items():
        print(f"{key} : {value}")
        
print_details(name= "Ahmet",age=18,city="Balıkesir")

```

```python
def print_details(name= "Misafir" , **info):
    print(f"name: {name}")
    for key, value in info.items():
        print(f"{key} : {value}")
        
print_details(name= "Ahmet",age=18,city="Balıkesir")
```
