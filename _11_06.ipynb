```python
# Cümlede En Sık Tekrarlanan Karakterler

# kullanıcıdan bir cümle al 

sentence = input ("Bir Cümle Giriniz: ")
print (sentence)

# Boşlukları Kaldır 

sentence = sentence.replace(" ","")
print(sentence)

#Frekans Bilgisi
frequency = {}   # key - value  

# Cümledeki her karakteri kontrol et 
for char in sentence:
    if char in frequency:
        frequency[char] += 1
    else:
        frequency[char] = 1
        
# en çok tekrar eden elemanı bul

most_frequency_char = ' '
max_count = 0

# bütün elemanlar (key - value ) dolaşma 
for  key, value in frequency.items():
    if value > max_count:
        max_count = value
        most_frequency_char = key
        
# sonuçları yazdır. 

print (frequency)
print('* '*10)
print(f"En Sık tekrar Eden karakter {most_frequency_char} ve tekrar değeri {max_count}")    

```
