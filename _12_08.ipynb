```python
#Alışveriş Sepeti Düzenleme

# girişi al
items = input ("Lütfen Alışveriş Listesindeki Ürünleri Giriniz (virgülle ürünleri ayırın)")

# ürünlistesi oluştur
items_list = items.split(',')

# temiz liste
cleaned_list = []

# ürünleri temizle 
for item in items_list:
    cleaned_list.append(item.strip().lower())

# sıralama 
sorted_list =sorted(cleaned_list)

# çıktı 
for item in sorted_list:
    print(f"{item.title()}")


```
