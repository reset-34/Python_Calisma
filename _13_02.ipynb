```python
numbers = [10, 15, 22, 33, 40, 55]
filtered_numbers = list (filter(lambda x: x>30, numbers))
print(filtered_numbers)
```

```python
from functools import reduce
numbers = list(range(1,6))
print (numbers)
product = reduce (lambda x, y : x*y , numbers)
print (product)
```

```python
names = ["Ali","aylin","Fatma","can"]
result = list (map (lambda name : name.upper(),names))
print(result)
```

```python
names = ["Ali","aylin","Fatma","can"]
result = list (map (lambda name : name.upper(),filter (lambda name: name.islower(),names)))
print(result)
```

```python
names = ["Ali","aylin","Fatma","can"]
result = list (map (lambda name : name.title(),filter (lambda name: name.islower(),names)))
print(result)
```

```python
numbers = list (range (1,11))
print(numbers)
even_numbers = list (filter(lambda x: x%2==0,numbers))
odd_numbers = list (filter (lambda x: x%2==1,numbers))

print("Çift sayılar", even_numbers)
print("Tek sayılar",odd_numbers)
```

```python
from functools import reduce
words = ["merhaba", "dünya","python"]
lengths = list(map (lambda w: len(w),words))
total_lengths = reduce(lambda x,y : x+y, lengths)

print (words)
print (lengths)
print (total_lengths)
```
