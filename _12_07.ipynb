```python
# Kelime Analizi (Toplam Kelime Sayısı, Toplam Harf Sayısı, En Uzun Kelime, En Uzun Kelime Sayısı )

# Giriş
text = "BTK AKademi İle Programlama öğreniyorum."

# Ayırma
word_list = text.split()

# Kelime Sayısı 
word_count = len (word_list)

# Harf Sayısı 
letter_count = 0
for char in text:
    if char.isalpha():
        letter_count+=1 

# En Uzun Kelime
longest_word = ""
for word in word_list:
    if len(word) > len (longest_word):
        longest_word=word




print ("Kelime Sayısı:", word_count)
print ("Harf Sayısı:", letter_count)
print ("En Uzun Kelime :", longest_word)
print ("En Uzun Kelimenin Uzunluğu :", len(longest_word))
```
