```python
toplam = 0

print ("Pozitif sayılar giriniz. (Bitirmek İçin 0 girin)")

while True:
    sayi = int (input ("Sayı"))
    if sayi==0:
        break
    elif sayi>0:
        toplam += sayi
    else:
        print("Negatif sayı yok sayıldı.")
print(f"Girilen pozitif sayıların toplamı : {toplam}")
    
    

```
