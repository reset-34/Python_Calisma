```python
# Girilen 5 Sayının ortalamasını alan program 

sayilar = []

for i in range (5):   # 5 sayı yeri al
    s= int (input("Bir sayı giriniz: "))
    sayilar.append(s)  # s değerini diziye ekle
    
toplam = 0
for sayi in sayilar:
    toplam += sayi 
    
ortalama = toplam / len (sayilar)
print (f"Girilen Sayıların Ortalaması:{ortalama:.2f}")   # sayının 2 haneli noktadan sonrası olsun
 
```
