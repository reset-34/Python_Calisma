```python
n= int (input ("Boyut için n değeri giriniz :"))
diyagonel = []
matris = [[0 for i in range(n)]for j in range (n)]
c = 1
for i in range(n):
    
    for j in range(n):
     matris [i][j] = c
     c+=1
     if i==j:
         diyagonel.append(matris[i] [j])
    
print (matris)
print (diyagonel)
```
