```python
n, m, c = 3,3,3

image = [[[0] *m for _ in range (n)] for _ in range (c)] #üç boyutlu dizi elde ettik

 #atama
# image [0] [0] [0] = 255
# image [1] [1] [1] = 255    # ctrl k c yorum satırı yapar 
# image [2] [2] [2] = 255
 
for i in range(n):
    image [i] [i] [i] = int (input (f"[{i}] [{i}] [{i}] için değer giriniz (0-255) "))
    
#gezinme 

for derinlik in range (c):
    for satır in range(n):
        for sutun in range(m):
            print(f"{image[satır][sutun][derinlik]: 4d}", end= ' ')
        print(" ")
    print(" ")
```
