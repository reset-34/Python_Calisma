```python
# Matris Kullanarak Ortalama Not Hesaplama 

#Notlar matrisi açalım
#giriş
grades_matris = [[70,80],
                 [90,85],
                 [60,75]]

# Ortalamaları Tutacağız
#çıkış
averages = []

for student_grades in grades_matris:
    average = sum (student_grades)/len (student_grades)
    averages.append(average)
    
print("Öğrenci Not Ortalamaları:",averages)
```
