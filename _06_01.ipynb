```python
sayi = float (input ("Bir sayı giriniz"))

if sayi>0:
    print (f"Girilen Sayi: {sayi:.2f}, pozitiftir.")
elif sayi<0:
    print (f"Girilen Sayi: {sayi:.2f}, negatiftir.")
else:
    print (f"Girilen Sayi: {sayi:.2f}, sıfırdır.")
```
