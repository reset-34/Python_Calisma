```python
# Hem 3'e hem 5'e Bölünebilen Sayılar

# başlangıç listesi 

numbers = [10,15,30,7,41,45,55]

# Hem 3'e hem 5'e Bölünebilen Sayı Listesi 
divisible_by_3and_5 = []

# listede döngü tanımı
for number in numbers:
    if number %3 == 0 and number%5 == 0:
        divisible_by_3and_5.append(number)
        
# Sonuçları Kullanıcıya Gösterme 
print ("# Hem 3'e hem 5'e Bölünebilen Sayılar: ",divisible_by_3and_5)

```
