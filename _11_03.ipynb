```python
# tekrar eden elemanları bulma 

numbers = [2,3,2,5,3,6]

# tekrarlı liste
repeat_items = []

for i in range (len (numbers)):
    count = 0
    # her bir sayı için çevrim 
    for j in range (len(numbers)):
        if numbers[i] == numbers[j]:
            count += 1
    # sayı tekrar edecek ve repeat_items da olmayacak 
    if count>1 and numbers[i] not in repeat_items:
        repeat_items.append(numbers[i])
print(" Tekrar Eden Sayılar : " , repeat_items)          


```
